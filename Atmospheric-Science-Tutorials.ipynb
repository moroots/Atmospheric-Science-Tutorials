### Introduction

This repository is a curated series of notebooks that build practical data analysis skills for atmospheric science using observational datasets. It starts with fundamentals (data access, quality control, units, and visualization) and advances to methods such as statistical inference, time-series analysis, and spatial analysis.

This introductory notebook provides the scope, organization, and conventions used throughout the series. Each notebook is designed to be self-contained, reproducible, and focused on real-world observations. Modeling workflows are intentionally out of scope.

Source code implementations are not included in these tutorials. The underlying tools and reusable code live in the `atmoz` package, which the notebooks call into as needed.

What you can expect from the series:

- Clear learning goals and short, focused workflows.
- Emphasis on reproducibility and transparent analysis.
- Examples that bridge observational data and scientific interpretation.

By the end of the series, you should be comfortable moving from raw atmospheric observations to defensible, publication-ready insights.
